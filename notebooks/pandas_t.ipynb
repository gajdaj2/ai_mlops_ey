```python
import pandas as pd

# Wczytanie danych
data = pd.read_csv('../data/Churn_Modelling_Spain.csv')

```

```python
# Wyświetlanie pierwszych i ostatnich wierszy
data.head(100)  # Pierwsze 5 wierszy
```

```python
print(data.tail())  # Ostatnie 5 wierszy
```

```python
# Informacje o danych
print(data.info())
```

```python

# Statystyki opisowe
print(data.describe())
```

Filtorwanie Danych

```python
# Wybór konkretnych kolumn
selected_columns = data[['CustomerId', 'CreditScore', 'Exited']]
print(selected_columns.head())
```

```python
# Filtrowanie na podstawie warunków (np. klienci, którzy odeszli)
churned_customers = data[data['Exited'] == 1]
churned_customers.head()
```

```python
# Klienci z wysokim saldem (>100,000) i aktywni
high_balance_active = data[(data['Balance'] > 100000) & (data['IsActiveMember'] == 1)]
high_balance_active.head()
```

Przekształcanie danych

```python
# Dodanie nowej kolumny - różnica wieku od średniego wieku
data['AgeDifference'] = data['Age'] - data['Age'].mean()

# Podział na grupy wiekowe
data['AgeGroup'] = pd.cut(data['Age'], bins=[18, 30, 45, 60, 80], labels=['18-30', '31-45', '46-60', '60+'])
data[['Age', 'AgeGroup']]
```

Agregracja i grupowanie

```python
# Średni wynik kredytowy według płci
credit_by_gender = data.groupby('Gender')['CreditScore'].mean()
credit_by_gender
```

```python
# Liczba klientów według grup wiekowych
age_group_counts = data['AgeGroup'].value_counts()
print(age_group_counts)
```

```python
# Średni balans konta według tego, czy klient odszedł
balance_by_exit = data.groupby('Exited')['Balance'].mean()
balance_by_exit
```

```python
# Pivot table: Średnie saldo według płci i członkostwa aktywnego
pivot_table = data.pivot_table(values='Balance', index='Gender', columns='IsActiveMember', aggfunc='mean')
pivot_table
```

```python
# Sortowanie danych według kilku kolumn
sorted_data = data.sort_values(by=['CreditScore', 'Balance'], ascending=[False, True])
sorted_data.head()
```

Wizualizacja danych

```python
import matplotlib.pyplot as plt

# Histogram wieku klientów
data['Age'].plot(kind='hist', bins=20, title='Rozkład wieku klientów')
plt.xlabel('Wiek')
plt.show()

# Wykres kołowy udziału płci
data['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Udział płci')
plt.ylabel('')
plt.show()
```

```python
churned_customers.to_csv('../data/churned_customers.csv', index=False)

# Zapis tabeli przestawnej
pivot_table.to_csv('../data/pivot_table.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Rozkład wieku klientów
sns.histplot(data=data, x='Age', kde=True, bins=20, color='blue')
plt.title('Rozkład wieku klientów')
plt.xlabel('Wiek')
plt.ylabel('Liczba klientów')
plt.show()
```

```python
# Rozkład balansu konta według płci
sns.boxplot(data=data, x='Gender', y='Balance', palette='pastel')
plt.title('Rozkład balansu konta według płci')
plt.xlabel('Płeć')
plt.ylabel('Saldo konta')
plt.show()
```

```python
# Relacja między wiekiem a wynikiem kredytowym
sns.scatterplot(data=data, x='Age', y='CreditScore', hue='Gender', palette='Set2')
plt.title('Zależność między wiekiem a wynikiem kredytowym')
plt.xlabel('Wiek')
plt.ylabel('Wynik kredytowy')
plt.legend(title='Płeć')
plt.show()
```

```python
sns.boxplot(data=data, x='IsActiveMember', y='NumOfProducts', palette='coolwarm')
plt.title('Aktywność członka a liczba produktów')
plt.xlabel('Aktywny członek (0 = Nie, 1 = Tak)')
plt.ylabel('Liczba produktów')
plt.show()
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa cieplna korelacji')
plt.show()
```

```python
sns.countplot(data=data, x='Geography', hue='Gender', palette='viridis')
plt.title('Podział klientów, którzy odeszli, według płci i regionu')
plt.xlabel('Region')
plt.ylabel('Liczba klientów')
plt.legend(title='Płeć')
plt.show()
```

```python
# Wynik kredytowy według statusu odejścia
sns.violinplot(data=data, x='Exited', y='CreditScore', palette='muted', split=True)
plt.title('Wynik kredytowy klientów według statusu odejścia')
plt.xlabel('Czy klient odszedł? (0 = Nie, 1 = Tak)')
plt.ylabel('Wynik kredytowy')
plt.show()
```

```python
# Grupy wiekowe
data['AgeGroup'] = pd.cut(data['Age'], bins=[18, 30, 45, 60, 80], labels=['18-30', '31-45', '46-60', '60+'])

# Średni balans w grupach wiekowych
sns.barplot(data=data, x='AgeGroup', y='Balance', estimator='mean', ci=None, palette='cool')
plt.title('Średni balans konta w grupach wiekowych')
plt.xlabel('Grupa wiekowa')
plt.ylabel('Średni balans')
plt.show()
```

```python
# Proporcja odejść według płci
sns.barplot(data=data, x='Gender', y='Exited', estimator=lambda x: sum(x)/len(x), ci=None, palette='Set1')
plt.title('Proporcja klientów, którzy odeszli, według płci')
plt.xlabel('Płeć')
plt.ylabel('Proporcja odejść')
plt.show()
```

```python
# Wskaźnik odejścia klientów według regionu
sns.barplot(data=data, x='Geography', y='Exited', estimator=lambda x: sum(x)/len(x), ci=None, palette='autumn')
plt.title('Wskaźnik odejścia klientów według regionu')
plt.xlabel('Region')
plt.ylabel('Proporcja odejść')
plt.show()
```
