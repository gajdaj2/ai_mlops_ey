```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Histogram
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title('Rozkład wartości rachunku')
plt.show()
```

```python
# Boxplot
sns.boxplot(data=tips, x='day', y='total_bill', palette='pastel')
plt.title('Rozrzut rachunków w poszczególne dni')
plt.show()
```

```python
# Scatterplot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', palette='cool')
plt.title('Zależność między rachunkiem a napiwkiem')
plt.show()
```

```python
# Lineplot
sns.lineplot(data=tips, x='size', y='total_bill', hue='day', marker='o')
plt.title('Średni rachunek w zależności od liczby osób przy stole')
plt.show()
```

```python
# Barplot
sns.barplot(data=tips, x='day', y='total_bill', estimator='mean', ci='sd', palette='muted')
plt.title('Średni rachunek w poszczególne dni')
plt.show()
```

```python
# Wybór tylko kolumn numerycznych
numeric_data = tips.select_dtypes(include=['float64', 'int64'])

# Korelacja między zmiennymi numerycznymi
correlation_matrix = numeric_data.corr()

# Mapa cieplna korelacji
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelacja między zmiennymi numerycznymi')
plt.show()
```

```python
# Pairplot - wiele wykresów na raz
sns.pairplot(tips, hue='sex', palette='husl')
plt.show()
```

```python
# Ustawianie stylu
sns.set_style('whitegrid')
sns.boxplot(data=tips, x='day', y='total_bill', palette='pastel')
plt.title('Boxplot z nowym stylem')
plt.show()
```

```python
# Kolory za pomocą palet
sns.histplot(data=tips, x='total_bill', kde=True, color='purple')
plt.title('Histogram z niestandardowym kolorem')
plt.show()
```
